```python
# Function to sample and display a few images from the directory
def sample_and_display_images(directory, sample_size=5):
    image_files = [f for f in os.listdir(directory) 
                   if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    # Sample randomly if we have more images than our sample size
    if len(image_files) > sample_size:
        import random
        image_files = random.sample(image_files, sample_size)
    
    # Display each image
    plt.figure(figsize=(15, 10))
    for i, img_file in enumerate(image_files, 1):
        img_path = os.path.join(directory, img_file)
        img = ImageDataAnalyzer.load_image(img_path)
        
        plt.subplot(1, len(image_files), i)
        plt.imshow(img)
        plt.title(f"{img_file}\n{img.size[0]}x{img.size[1]}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Sample and display some images
sample_and_display_images(image_directory)
```

```python
# Collect image sizes
print("Collecting image file sizes...")
image_sizes = []
image_names = []

for image_file in os.listdir(image_directory):
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(image_directory, image_file)
        file_size = os.path.getsize(image_path) / 1024  # Size in KB
        image_sizes.append(file_size)
        image_names.append(image_file)

# Create DataFrame for analysis
size_df = pd.DataFrame({
    'Filename': image_names,
    'Size_KB': image_sizes
})

# Sort by file size
size_df = size_df.sort_values('Size_KB', ascending=False).reset_index(drop=True)

# Display top 10 largest images
print("\nTop 10 largest images:")
print(size_df.head(10))

# Plot distribution of file sizes
plt.figure(figsize=(12, 6))
sns.histplot(size_df['Size_KB'], bins=30, kde=True)
plt.title('Distribution of Image File Sizes')
plt.xlabel('File Size (KB)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Function to analyze color distribution of an image
def analyze_image_colors(image_path):
    img = ImageDataAnalyzer.load_image(image_path)
    img_array = np.array(img)
    
    # Check if image is RGB
    if len(img_array.shape) == 3 and img_array.shape[2] >= 3:
        # Get RGB channels
        r = img_array[:,:,0].flatten()
        g = img_array[:,:,1].flatten()
        b = img_array[:,:,2].flatten()
        
        # Plot histograms for each channel
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.hist(r, bins=50, color='red', alpha=0.7)
        plt.title('Red Channel')
        plt.xlabel('Pixel Value')
        plt.ylabel('Frequency')
        
        plt.subplot(1, 3, 2)
        plt.hist(g, bins=50, color='green', alpha=0.7)
        plt.title('Green Channel')
        plt.xlabel('Pixel Value')
        
        plt.subplot(1, 3, 3)
        plt.hist(b, bins=50, color='blue', alpha=0.7)
        plt.title('Blue Channel')
        plt.xlabel('Pixel Value')
        
        plt.tight_layout()
        plt.show()
        
        # Display the image
        ImageDataAnalyzer.display_image(img, title=os.path.basename(image_path))
        
        return {
            'mean_red': np.mean(r),
            'mean_green': np.mean(g),
            'mean_blue': np.mean(b),
            'std_red': np.std(r),
            'std_green': np.std(g),
            'std_blue': np.std(b)
        }
    else:
        print(f"Image {os.path.basename(image_path)} is not RGB")
        return None

# Analyze color distribution for a sample image
if image_count > 0:
    sample_image = os.path.join(image_directory, image_names[0])
    print(f"Color analysis for sample image: {image_names[0]}")
    color_stats = analyze_image_colors(sample_image)
    if color_stats:
        print("Color statistics:")
        for key, value in color_stats.items():
            print(f"{key}: {value:.2f}")
```

```python
import os
import numpy as np
import pandas as pd
from collections import Counter
from PIL import Image

# Folder path
image_directory = r"meningioma"

# Verify the path exists
if not os.path.exists(image_directory):
    print(f"Error: Directory '{image_directory}' not found!")
    exit()

# Store image properties
image_data = []

# Process images
for file in os.listdir(image_directory):
    if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
        filepath = os.path.join(image_directory, file)
        
        try:
            with Image.open(filepath) as img:
                width, height = img.size
            file_size_kb = os.path.getsize(filepath) / 1024
            aspect_ratio = width / height if height > 0 else 0
            image_data.append([file, width, height, aspect_ratio, file_size_kb])
        except Exception as e:
            print(f"Error processing {file}: {e}")

# Create DataFrame
size_df = pd.DataFrame(image_data, columns=["Filename", "Width", "Height", "Aspect_Ratio", "Size_KB"])

# Debug: Print DataFrame columns
print("Columns in size_df:", size_df.columns)

# Handle empty DataFrame
if size_df.empty:
    print("No valid images found! Exiting...")
    exit()

# Sort by size
size_df = size_df.sort_values(by="Size_KB", ascending=False)

# Extract values
aspect_ratios = size_df["Aspect_Ratio"].tolist()
image_sizes = size_df["Size_KB"].tolist()

# Summary Report
print("\n=== IMAGE FOLDER ANALYSIS SUMMARY ===")
print(f"Total images analyzed: {len(size_df)}")
print(f"Average image dimensions: {size_df['Width'].mean():.1f} x {size_df['Height'].mean():.1f} pixels")
print(f"Average aspect ratio: {np.mean(aspect_ratios):.2f}")
print(f"Average file size: {np.mean(image_sizes):.2f} KB")
print(f"Smallest image: {size_df.iloc[-1]['Filename']} ({size_df['Size_KB'].min():.2f} KB)")
print(f"Largest image: {size_df.iloc[0]['Filename']} ({size_df['Size_KB'].max():.2f} KB)")

# Identify common aspect ratios
rounded_ratios = [round(r, 2) for r in aspect_ratios]
common_ratios = Counter(rounded_ratios).most_common(5)

print("\nMost common aspect ratios:")
for ratio, count in common_ratios:
    print(f"  {ratio:.2f}: {count} images ({count/len(aspect_ratios)*100:.1f}%)")
```
